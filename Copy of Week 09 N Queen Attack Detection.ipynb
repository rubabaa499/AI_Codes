# N Queen Attack Detection

```python
def drawBoard(P):
  N = len(P)
  for i in range(1,N+1):
    print("|", end="")
    for j in range(1,N+1):
      if (i,j) in P:
        ind = str(1+ P.index((i,j)))
        print(" "+ind+" |", end= "")
      else:
        print(" * |", end= "")
    print()
```

```python
from math import ceil
import random
def generate_positions(N):
    CL = list(range(0,N*N));
    P = []
    for i in range(N):
        p = random.choice(CL)
        CL.remove(p)
        r = ceil((p+1)/N)
        c = 1+p%N
        P.append((r,c))
        print('Q{}: {},{}'.format(i+1,r,c))
    return P

p = generate_positions(4)
print(p)
```

```python
N = 5
positions = generate_positions(N)
drawBoard(positions)
```

```python
# row attack
def rowAttacK(P):
  N = len(P)
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if P[i][0] == P[j][0]:
        print('row attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j]) 
        count = count + 1 
  return count

rowAttacK(positions)   
```

```python
#column attack
def columnAttack(P):
    N = len(P)
    count = 0
    for i in range(N):
        for j in range(i+1,N):
            # check if the queens are in the same column
            if P[i][1] == P[j][1]:
                # print('column attack:', end = " " )
                # print(P[i], end = ", ")
                # print(P[j]) 
                count = count + 1 
    return count
```

```python
# diagonal atack
def diagonalAttack(P):
    N = len(P)
    count = 0
    # for each queen
    for i in range(N):
        # for each other queen
        for j in range(i+1,N):
            # check if the queens are in the same diagonal
            if abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):
                # print('diagonal attack:', end = " " )
                # print(P[i], end = ", ")
                # print(P[j]) 
                count = count + 1 
    return count
```

```python
# implement a method to print a state is valid or not
# input row and col (two lists)

def isValid(P):
    if rowAttacK(P) + columnAttack(P) + diagonalAttack(P) == 0:
        return True
    else:
        return False
```

```python
# You can also repeat to get valid positions [Random search]
N = 5
positions = generate_positions(N)
valid = isValid(positions)
print('validity:',valid)
print('generated positions->',positions)

while not valid:
    positions = generate_positions(N)

    print('generated positions->',positions)
    valid = isValid(positions)
    print('validity:',valid)
drawBoard(positions)
```
